```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
import numpy as np
from sklearn import datasets
```

```python
iris=datasets.load_iris()
x=iris.data
y=iris.target
```

```python
print('sepal-length','sepal-width','petal-length','petal-width')
print(x)
```

```python
print('class: 0-Iris-Setosa, 1- Iris-Versicolour, 2- Iris-Virginica')
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Plotting decision boundaries for two features (sepal length and sepal width)
x_train_2d = x_train[:, :2]
x_test_2d = x_test[:, :2]

classifier_2d = KNeighborsClassifier(n_neighbors=5)
classifier_2d.fit(x_train_2d, y_train)

# Create a meshgrid for plotting decision boundaries
x_min, x_max = x_train_2d[:, 0].min() - 1, x_train_2d[:, 0].max() + 1
y_min, y_max = x_train_2d[:, 1].min() - 1, x_train_2d[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = classifier_2d.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot training points
plt.scatter(x_train_2d[:, 0], x_train_2d[:, 1], c=y_train, cmap=cmap_bold,
            edgecolor='k', s=20)
plt.scatter(x_test_2d[:, 0], x_test_2d[:, 1], c=y_test, cmap=cmap_bold,
            edgecolor='k', s=100, marker='x')

plt.title("K-NN decision boundaries (Iris dataset - Sepal features)")
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
import seaborn as sns

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", 
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Confusion Matrix for K-NN Classifier")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
print("Accuracy Metrics")
print(classification_report(y_test,y_pred))
```

```python
print("Predictions :")
for i in range(len(y_test)):
    print(f"Sample {i+1}: True Label: {y_test[i]}, Predicted Label: {y_pred[i]}, {'Correct' if y_test[i] == y_pred[i] else 'Wrong'}")
```
